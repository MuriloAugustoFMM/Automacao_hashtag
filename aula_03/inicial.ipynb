```python
import pandas as pd
#import scikit_learn as sk
# carregar tabela
tabela = pd.read_csv('clientes.csv')
```

```python
# tratar os dados da tabela alterando o tip de dados
from sklearn.preprocessing import LabelEncoder

# codificador de profissão
codificardor_profissao = LabelEncoder()
tabela["profissao"] = codificardor_profissao.fit_transform(tabela["profissao"])

# codificador de mix_credito
codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"])

# codificador de comportamento
codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```
